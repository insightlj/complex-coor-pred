```python
import sys
sys.path.append("/home/rotation3/complex-coor-pred/")

import h5py
import torch
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

train_file = h5py.File("AlignCoorConfusion/h5py_data/train_dataset.h5py", "r")

```

```python
index = 10
for index in [0,10,100,200,500,1000,2000,5000,10000,15000,20000]:
    lddt_score = torch.from_numpy(np.array(train_file["protein"+str(index)]["lddt_score"], dtype=np.float32)).squeeze()
    fig, ax =plt.subplots(1,2,constrained_layout=True, figsize=(8, 3))
    sns.heatmap(lddt_score,cmap="RdBu_r",ax=ax[0])
    lddt_score.mean()
    lddt_per_chain = lddt_score.mean(dim=1)
    lddt_per_residuse = lddt_score.mean(dim=0)
    sns.lineplot(lddt_score[0],ax=ax[1])
    sns.lineplot(lddt_per_residuse,ax=ax[1])
    plt.show()
    print("lddt_score_mean", lddt_per_residuse.mean())
    print("lddt_score", lddt_score[0].mean())
    print("lddt_score_max", (lddt_score.max(dim=0)[0]).mean())
    print("_________________________")


### 这说明，预测的结果是有很大的提升空间的。只需要预测每两套坐标系之间的位置
```
