```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import torch

f = open("/home/rotation3/complex-coor-pred/utils/AlignCoorSample/log_gate_train.txt", "r")

protein_index = []
initial_lddt = []
final_lddt = []
expected_lddt = []

for line in f:
    line = line.strip()
    information = line.split()
    if len(information)<2 or information == ['final', 'lddt:']:
        continue
    if information[0] == "protein":
        protein_index.append(eval(information[1]))
    elif information[0] == "initial":
        initial_lddt.append(eval(information[2]))
    elif information[0] == "final":
        final_lddt.append(eval(information[2]))
    elif information[0] == "ideal":
        expected_lddt.append(eval(information[2]))

initial_lddt = np.array(initial_lddt)[:-1]
final_lddt = np.array(final_lddt)
expected_lddt = np.array(expected_lddt)
print(np.mean(initial_lddt), np.mean(final_lddt), np.mean(expected_lddt))


sorted_initial_lddt = np.sort(initial_lddt)
index4sort = np.argsort(initial_lddt)
aligned_final_lddt = final_lddt[index4sort]
aligned_excepted_lddt = expected_lddt[index4sort]
sns.lineplot(sorted_initial_lddt)
sns.lineplot(aligned_final_lddt)
sns.lineplot(aligned_excepted_lddt)


```
